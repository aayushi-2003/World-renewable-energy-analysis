```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
pd.set_option('display.max.rows', 2270)
```

```python
df = pd.read_excel(r'/Users/mourya/Desktop/Data Analytics Course Project/IndiaGDPvsEverything.xlsx')

df
```

```python
df = df.drop(columns = ['Solar_Generation_TWh', 'Hydro_Generation_TWh', 'Wind_Generation_TWh', 'Geo_Biomass_Other_TWh'])
```

```python
df.corr()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.heatmap(df.corr(), annot=False, cmap='YlGnBu', vmin=0.5, vmax=1)  # Disable automatic annotations

ax.set_xticklabels(ax.get_xticklabels(), fontsize=16)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=16)

# Iterate over each cell and add annotation manually
for (i, j), z in np.ndenumerate(df.corr()):
    # Check if value exceeds the threshold
    if z >= 0.8:  
        text_color = "white"  # Lighter color for high correlation values
    else:
        text_color = "black"  # Default color for other squares

    ax.text(j + 0.5, i + 0.5, "{:.3f}".format(z), ha='center', va='center', fontsize=18, color=text_color)

plt.show()
```
